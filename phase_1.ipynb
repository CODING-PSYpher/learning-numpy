# learning about numpy

```python
import numpy as np
import time
```

## creating array from list

```python
array_1d = np.array([1,2,3,4])
print("1D array: ", array_1d)
array_2d = np.array([[1,2,3],[4,5,6]])
print("2D array: ", array_2d)
```

### list vs numpy array

```python
py_list = [1,2,3,4]
print("Python list multiplication: ", py_list * 2)
np_array= np.array([1,2,3,4]) #element wise multiplication
print("numpy array multiplication: ", np_array * 2)


start1 = time.time()
py_list2 = [i*2 for i in range(10000000)]
print("\n List operation time: ", time.time()-start1)
start2 = time.time()
np_array2= np.arange(10000000) * 2
print("\n List operation time: ", time.time()-start2)
```

### creating array from scratch

```python
zeros= np.zeros((3,3))
print("zeros array: \n", zeros)
ones= np.ones((3,3))
print("ones array: \n", ones)
full= np.full((3,3), 69)
print("full array: \n", full)
random= np.random.random((3,3))
print("random array: \n",random)
sequence= np.arange(0,11, 2)
print("sequence array: \n",sequence)
```

### Matrice, Vector and Tensor

```python
vector= np.array([1,2,3])
print("vector: \n", vector)
matrix= np.array([[1,2,3],
                   [4,5,6]])
print("matrix: \n", matrix)
tensor= np.array([[[1,2],[3,4]],
                  [[5,6],[7,8]]])
print("tensor: \n", tensor)
```

Array props

```python
arr= np.array([[1,2,3],
               [4,5,6]])

print("shape: ", arr.shape)
print("dimension: ", arr.ndim)
print("size: ", arr.size)
print("data types: ", arr.dtype)
```

###Array reshaping

```python
arr = np.arange(9)
print("original array: ", arr)
reshaped= arr.reshape((3,3))
print("\n reshaped array: \n", reshaped)
flatten= reshaped.flatten()
print("\n flatten array: ", flatten)
raveled= reshaped.ravel() #it returns view insted of copy, so original is safe
print("\n raveled array: ", raveled)

transpose = reshaped.T
print("\n transopose array: ", transpose)
```
