```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("=== Zomato Sales Data Analysis === ")
print("\n Sales data shape: ", sales_data.shape)
print("\n Sample data for 1st 3 restaurant: \n", sales_data[:3])
```

```python
#total sales per year

print(np.sum(sales_data, axis=0))
yearly_total= np.sum(sales_data[:, 1:], axis=0)
print("\n Total sales per year: ", yearly_total)

#minimum sales per year
mini_sale = np.min(sales_data[:, 1:], axis=1)
print("\n Minimum sales per year: ", mini_sale)
#maximum sales per year
max_sale = np.max(sales_data[:, 1:], axis=1)
print("\n Maximum sales per year: ", max_sale)

#avg sales per restaurant

avg_sale= np.mean(sales_data[:, 1:], axis=1)
print("\n Average sales per rest: ", avg_sale)
```

```python
cumsum = np.cumsum(sales_data[:,1:], axis=1)
print(cumsum)


plt.figure(figsize=(8, 6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average cumlative sales per restaurant")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```
